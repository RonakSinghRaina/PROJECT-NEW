# Statistical Inference 

The statistical inference is an act of drawing conclusions or answering questions about an unknown “state of nature” based solely on theory and observation. What is that we can learn using the observed data?

In some cases, the scientists measure the properties of a limited sample of certain objects in order to understand and learn about the properties of the large underlying population of such objects (population).

We are doing statistical inference when 

- smooths over discrete observations to understand the underlying continuous phenomenon 
- seeks to quantify relationships between observed properties
- tests whether an observation agrees with an assumed theory
- divides a sample into subsamples with distinct properties
- tries to compensate for flux limits and non-detections
- investigates the temporal behavior of variable sources
- infers the evolution of the population by studying the objects at different stages 
- characterizes and models patterns in wavelength, images or space

## Two main aspects of inference are: estimation and the testing of hypotheses.

Regression, goodness-of-fit, classification and many other statistical procedures fall under its framework. 

Statistical inference can be parametric, nonparametric and semi-parametric. 

- **Parametric inference** requires that the scientist makes some assumptions regarding the mathematical structure of the underlying population, and this structure has parameters to be estimated from the data at hand. Example: Linear regression is an example of parametric inference; Pearson correlation coefficient.  

- **Nonparametric** procedures make no assumption about the model structure or the distribution of the population. Example: The Kolmogorov–Smirnov hypothesis test and the rank-based Kendall’s τ correlation coefficient, Spearman correlation coefficient. 

- **Semi-parametric** methods combine nonparametric and parametric procedures. Example: local regression models. 


# Point Estimation

If the shape of the probability distribution, or relationship between variables (model), of the underlying population is well-understood, then it remains to find the parameters of the distribution or relationship.

## The goal of estimating plausible or “best” values of $\theta$ (parameters) of the model based on observations is called point estimation

# Statistical Model 
Since an observation $x$ is an outcome of a random experiment we consider $x$ as a value (realization) of a random variable $X$. Its distribution is at least partly unknown. Statistical inference is concerned with methods of using these observation $x$ to obtain information about the distribution of $X$. That means starting point of the statistical analysis is not one probability measure on X, but a class of possible probability measures ${\cal{P}}$, parameterized by a parameter $\theta$:

${\cal{P}} = \{{\cal{P}_{\theta}}: \theta \in \Theta\} $

The set $\Theta$ is called the $\bf{parameter \, space}$. We call this class of probability measures on the sample space $X$ a $\bf{statistical\, model}$. Here ${\cal{P}}$ is a family of product measures:

${\cal{P}_{\theta}} = {\cal{P}_{1, \theta}} \otimes .... \otimes {\cal{P}_{n, \theta}} $

### A statistical model is a mathematical framework that describes how data is generated, using a probability distribution with parameters. 


A statistical model consists of:
- A set of random variables
- A probability distribution
- A set of unknown parameters

For example, if we assume that a dataset follows a normal distribution:
$
\begin{equation*}
X \sim \mathcal{N}(\mu, \sigma^2)
\end{equation*}
$
where:
- \( X \) is the random variable
- \( $\mu$ \) (mean) and \( $\sigma^2$ \) (variance) are unknown parameters or in practice, the mean is a function of the model parameters which are unknown and variance corresponds to the error of the data point. 

**A statistical model is a mathematical framework that describes how data could be generated through a family of probability distributions, but it does not contain the actual observed data.**

# Likelihood Function $({\cal{L}})$

Likelihood function is not probability but it is proportional to a probability. 

Likelihood function represents the joint probability of the observed data given a statistical model. 
In other words, the likelihood is defined as the total probability of observing the data given the model and the current parameters.

For a fixed observation $x$ of a random variable $X$ with a probability function $p(x| \theta)$ the likelihood function ${\cal{L}}(\theta| x)$ is defined by
     
# $ {\cal{L}}(\theta|x) = K \times p(x|\theta) $.

For a same given data, the K remains the same and therefore when it is a question of comparing Likelihood function it doesn't matter if it is there or not in the expression. A likelihood is not actually a probability it doesn’t obey various rules of probability; for example, likelihoods need not sum to 1.

A critical difference between probability and likelihood is in the interpretation of what is fixed and what can vary. In the case of a conditional probability, $P(Data|Hypothesis)$, the hypothesis is fixed and the data are free to vary. Likelihood, however, is the opposite. The likelihood of a hypothesis, ${\cal{L}}(H)$, conditions on the data as if they are fixed while allowing the hypotheses to vary.

Unlike a probability, a likelihood has no real meaning per se due to the arbitrary constant K. They are meaningful only when compared and therefore, likelihoods are inherently comparative. 



# Maximum Likelihood Estimate

Example:

- For data whose underlying source distribution is assumed to be Gaussian, the likelihood is 

# ${\cal{L}} =
\begin{equation*}
\prod_{i=1}^{N} \frac{1}{\sqrt{2\pi \sigma_i^2}} exp\left[\frac{-(y_i - m_i)^2}{2\sigma_i^2}\right]
\end{equation*}
$

where $y_i$ are the observed data, $\sigma_i$ their errors, and $m_i$ the values of the predicted data based on the model (with current parameters belonging to $\Theta$) and instrumental response.


Since this is a complicated probability product function, we can take the logarithm of the ${\cal{L}}$

# $ log({\cal{L(\Theta)}}) = \frac{-1}{2} \sum_{i=1}^{N} log(2\pi \sigma_i^2) - \sum_{i=1}^{N}\frac{ (y_i - m_i)^2}{2 \sigma_i^2}$


# $ 2 log({\cal{L(\Theta)}}) = -\left[ \sum_{i=1}^{N} \, log(2\pi \sigma_i^2) + \sum_{i=1}^{N} \frac{ (y_i - m_i)^2}{\sigma_i^2}\right] $

In order to find the parameters, $\Theta$ that maximises the likelihood, we take the first derivative of the left side of the above equation

# $ \frac{\partial{\, 2 log({\cal{L(\Theta)}})}}{\partial{\Theta}} = 0$

## For example, the model of my choice is a straight line such that model counts, $m = a x + b$. Then in the above equation $\Theta \equiv \{a,b\}$. 

This methodology gives you the Maximum Likelihood estimate. For some models (as shown above), these equations can be explicitly solved for $\Theta$ but in general no closed-form solution to the maximization problem is known or available, and an MLE can only be found via $\bf{numerical \, optimization}$.


Generally, **negative of log-likelihood** is taken, thereby most of the available algorithms for the likelihood function optimization find the minimum, not the maximum, of the function. Since the likelihood function is usually the product of many probabilities ($0 < {\cal{P}} <1$) they tend to be very small. Hence, it is much more convenient numerically to deal with the logarithm of the likelihood. Thus, most of the optimisation algorithms instead of finding the maximum of the likelihood, we find the minimum of the $-log({\cal{L}})$ function. 
**Thus, these optimisation algorithms are referred to as minimisers.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import minimize
import pylab as py

# An example of an empirical model: cutoff power law
def cutoff_powerlaw(E,K,index,ecutoff):
    # E in units of keV
    N = K * E**(index) * np.exp(-E/ecutoff)
    return N

# An example of another empirical model: a straight line
def line(E,m,c):
    N = m*E + c
    return N

# Model function 
n = 100 # no: of data points
E = np.linspace(1,10000,n)

# input values for 
m=2
c = 10

plt.figure()
plt.plot(E,line(E,m,c),'r-')
plt.plot(E,cutoff_powerlaw(E,10,-1.5,100.0),'b-')
plt.xlabel('E')
plt.ylabel('N')
plt.yscale('log')
plt.xscale('log')
plt.show()
  
```

```python
# Simulating fake dataset for a straight line fit
m_ = 2 #np.random.normal(2,0.05)
c_ = 15 #np.random.normal(20,5)

# Model function 
n = 100 # no: of data points
E = np.linspace(1,100,n)

N = line(E,m_,c_)
error = np.sqrt(N)  # Poisson error approximation
Data = N + np.random.normal(0, error)  
# Generate synthetic observed data by adding Gaussian noise 
# with variance equal to the Poisson variance (sqrt(N)).
# This approximates Poisson counting statistics with a Gaussian
# distribution, valid when counts are sufficiently large.
print('True value of parameters',m_, c_) 
```

```python
# Calculate negative log likelihood
# The current definition of likelihood is based on the Guassian statistical model.
def log_likelihood_line(params):
    N = line(E,params[0],params[1])
    LL = -np.sum(stats.norm.logpdf(Data, loc=N, scale=error))
    return LL
```

```python
m__=2
c__=14
initparams=[m__,c__]
results = minimize(log_likelihood_line,x0=initparams,method='Nelder-Mead',tol=1e-6)
print(results)
print(results.x)
#print(results.keys())

estParms = results.x
E_ = np.linspace(1,100,200)
N_est = line(E_,*estParms) 


plt.figure()
plt.errorbar(E,Data,yerr=np.sqrt(N),fmt='bo',alpha=0.5)
plt.plot(E_,N_est,'r-',linewidth=4)
# plt.yscale('log')
# plt.xscale('log')
plt.show()


#Plotting the likelihood
m_1 = np.linspace(0,10,500)
c_1 = np.linspace(1,50,500)
```

Near the maximum likelihood point, the negative log-likelihood function can be approximated by a quadratic expansion around the best–fit parameters $\hat{\theta}$. This approximation is

$$
- \ln \mathcal{L}(\theta) \approx - \ln \mathcal{L}(\hat{\theta})
+ \frac{1}{2} (\theta - \hat{\theta})^{T} H (\theta - \hat{\theta}),
$$

where $H$ is the **Hessian matrix** defined as

$$
H_{ij} = \frac{\partial^2 (-\ln \mathcal{L})}{\partial \theta_i \partial \theta_j}
\Bigg|_{\theta=\hat{\theta}} .
$$

The Hessian measures the **curvature of the likelihood surface** near the optimum: a steeper curvature implies smaller parameter uncertainties, while a flatter curvature implies larger uncertainties. The **covariance matrix of the parameters** is therefore approximated by the inverse Hessian,

$$
\mathrm{Cov}(\theta) \approx H^{-1}.
$$

The diagonal elements give the parameter variances $\sigma_i^2$, while the off–diagonal elements describe correlations between parameters.

# Goodness of Fit 

## How confident can one be that the observed data can have been produced by the best-fit model?

Goodness of fit tells you how well your model has fit the given set of observations. 


The first step would be look at the 

- ## Residuals:
The observed source spectrum is a sum of the source emission + noise. Thus, if the assumed model function, $\cal{M}$, is similar to the actual source emission, then the resultant residuals should be just noise, which would be random variations about zero. If the residuals show any particular pattern or deviations from zero in any direction, we can conclude that the assumed model function is incorrect. However, if the residuals appear random with no particular deviation patterns, we can say that the assumed model is consistent with the data. 

The residuals can be plotted as follows (data-model)/error where error is calculated as the square root of the model predicted number of counts.

## It is worth recognising that we can always reject a model with certainty, however, we can never be absolutely sure about a model. Instead we can only claim a model to be consistent with the observed data at certain level of confidence. The determination of this level of confidence is known as the ${\bf 'goodness \, of \, fit'}$.  



```python
def bb(E,K,T):
    #T=60
    N = (K*8.0525*E**2)/(T**4 * (np.exp(E/T) -1))
    return N

# def sin(E,K):
#     N = K*np.sin(E)
#     return N
    
popt_, pcov_ = curve_fit(bb, xdata,ydata,p0=[1.8,10],maxfev=10000)
print('The fit values are popt',popt_)
print('The covariance matrix',pcov_)
```

```python
plt.figure()
E_ = np.linspace(1,100,200)
plt.errorbar(E,Data,yerr=np.sqrt(N),fmt='bo',alpha=0.5)
plt.plot(E_,bb(E_,*popt_),'r-',linewidth=4)
plt.show()
```

```python
res = np.divide((Data - line(E,*popt)), np.sqrt(line(E,*popt)))
res_err = np.sqrt((np.sqrt(Data) / Data)**2 + ((np.sqrt(line(E,*popt)) / line(E,*popt)))**2)*np.abs(res)  

res_bb = np.divide((Data - bb(E,*popt_)), np.sqrt(bb(E,*popt_)))
res_err_bb =np.sqrt((np.sqrt(Data) / Data)**2 + ((np.sqrt(bb(E,*popt_)) / bb(E,*popt_)))**2)*np.abs(res_bb)  

plt.figure()
plt.errorbar(E,res,yerr=res_err,fmt='ko',alpha=0.5)
plt.errorbar(E,res_bb,yerr=res_err_bb,fmt='ro',alpha=0.5)
plt.ylim(-6,15)
plt.axhline(y=0)
plt.ylabel('(Data-Model)/(Error of model)')
plt.show()
```

- ## Reduced Chi-square: 
In $\chi^2$ statistics, the goodness of fit is determined by the parameter known as the reduced $\chi^2$ defined by 
# \begin{equation}
 \chi_{\nu}^2 = \frac{\chi^2}{\nu}
\end{equation}
where $\nu$ is degrees of freedom. Degrees of freedom is equal to no: of data points - no: of parameters in your model. 
If $\chi_{\nu}^2 \gg 1$ means that the fit has not been able to model data properly, in other words the error variance ($\sigma_i$) has been underestimated. If $\chi_{\nu}^2 \ll 1$, it means the model is over-fitting the data such that the $\sigma_i$ is overestimated. If $\chi_{\nu}^2 = 1$, then the model matches the data consistently within the errors. 


### Note: However, there are several reasons that $\chi_{\nu}^2$  can be large, and being a bad fit is only one of them; other possibilities include: defining $\chi_{\nu}^2 $ with a different denominator; underestimation of errors; not accounting for systematic errors; and applying it to data whose source distribution may not be Gaussian.


The chi-squared distribution ( $\chi^2$-distribution) with $k$ degrees of freedom is the distribution of a sum of the squares of $k$ independent standard normal random variables. The mean of the distribution is $k$ and variance is $2k$.



```python
def chi_square(data,error,model,no_parameters):
    chi_2 = round(np.sum((data-model)**2/error**2),2)
    dof = len(data)- no_parameters
    print(dof)
    reduced_chi_2 = round(chi_2/dof,2)
    return(chi_2,reduced_chi_2)


line_fit = chi_square(Data,error,line(E,*popt),2)
print(line_fit) 
bb_fit = chi_square(Data,error,bb(E,*popt_),2)
print(bb_fit) 
```

- ## Monte Carlo simulations:
In case of Cstat or Pgstat, their obtained value from the fit does not give us any measure of the goodness of fit, as the value depends on the number of measurements and values of the data points. Thus, in order to assess the goodness of fit, one has to perform Monte Carlo simulations of the spectra based on the best fit model and create a distribution of the statistic. Then note where the obtained statistic of the fit lie on the corresponding cumulative distribution plot, thereby giving the level of confidence at which the best fit to the data is obtained. 

## Compute goodness of fit by generating Monte Carlo datasets and fitting the current model on them. The fraction of synthetic datasets which have a value for the likelihood larger or equal to the observed one is a measure of the goodness of fit

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Monte Carlo Simulations to assess goodness-of-fit
num_simulations = 10000
red_chi_square_sim = []
chi_square_sim = []

for _ in range(num_simulations):
    # Generate synthetic data assuming Gaussian fluctuations
    simulated_error = np.sqrt(line(E, popt[0], popt[1])) #Poisson error 
    simulated_data = line(E, popt[0], popt[1]) + np.random.normal(0, simulated_error) # Gaussian approximation to Poisson
    
    # Fit simulated data
    sim_popt, _ = curve_fit(line, E, simulated_data, sigma=simulated_error, absolute_sigma=True)
    sim_m, sim_c = sim_popt 

    
    # Compute chi-square for simulated dataset
    no_parameters=2
    dof = len(simulated_data)- no_parameters
    sim_residuals = simulated_data - line(E, sim_m, sim_c)
    sim_chi_square = np.sum((sim_residuals / simulated_error) ** 2)
    red_chi_square_sim.append(sim_chi_square / dof)
    chi_square_sim.append(sim_chi_square)


## Comparing it with the mean of a chi-square distribution of k degrees of freedom. 
# Generate chi-square distributed values using Monte Carlo
num_samples = 100000  # Number of Monte Carlo samples
chi_square_samples = np.random.chisquare(df=dof, size=num_samples)

# Compute estimated mean
estimated_mean = np.mean(chi_square_samples)


# Theoretical mean of chi-square distribution
theoretical_mean = dof  # Mean of chi-square distribution is k

print(f"Estimated Chi-square dist Mean from Simulation: {estimated_mean:.5f}")
print(f"Theoretical Chi-square dist Mean: {theoretical_mean}")
print(f"Chi-square estimate from fitting: {line_fit[0]}")
print(f"Degrees of freedom: {dof}")
print(f"Reduced Chi-square estimate from the fit: {line_fit[0]/dof}")

# Compare observed reduced chi-square with simulated distribution
# plt.hist(red_chi_square_sim, bins=30, alpha=0.7, color='b', label="Simulated Reduced Chi-Square")
# plt.axvline(line_fit[1], color='r', linestyle='dashed', linewidth=2, label="Observed Reduced Chi-Square")
#plt.axvline(line_fit[1], color='r', linestyle='dashed', linewidth=2, label="Observed Reduced Chi-Square")

plt.hist(chi_square_sim, bins=30, alpha=0.7, color='b', label="Simulated Chi-Square")
plt.axvline(line_fit[0], color='r', linestyle='dashed', linewidth=2, label="Observed Chi-Square")
plt.axvline(estimated_mean, color='g', linestyle='dashed', linewidth=2, label="Expected mean of Chi-square distribution")
plt.xlabel("Chi-Square")
plt.ylabel("Frequency")
plt.legend()
plt.title("Monte Carlo Simulations: Goodness-of-Fit")
plt.show()

```

```python
# Convert to numpy arrays
chi_square_sim = np.array(chi_square_sim)

# Observed chi-square
chi2_obs = line_fit[0]

# Goodness-of-fit estimate (p-value)
p_value = np.mean(chi_square_sim >= chi2_obs)

print(f"Goodness-of-fit estimate (p-value): {p_value:.4f}")

# Sort chi-square values for cumulative plot
sorted_chi2 = np.sort(chi_square_sim)

# Fraction of simulations with chi2 greater than each value
fraction_greater = 1.0 - np.arange(1, len(sorted_chi2)+1) / len(sorted_chi2)

plt.figure()

plt.plot(sorted_chi2, fraction_greater, label="Monte Carlo cumulative fraction")

plt.axvline(chi2_obs, linestyle='--', linewidth=2,
            label="Observed Chi-square")

plt.xlabel("Chi-square")
plt.ylabel("Fraction with $\chi^2$ greater than x")
plt.title("Cumulative Monte Carlo Goodness-of-Fit")
plt.legend()

plt.show()
```

### The role of goodness-of-fit is diagnostic: it tells you whether something looks statistically unusual.

#### The goodness-of-fit does not measure how good a model is, but rather whether the data are statistically compatible with the model and its assumed uncertainties. Extreme values indicate a problem; moderate values simply indicate consistency.

## A model is considered a good fit if the observed deviations between the data and the model are statistically consistent with the expected noise, typically reflected by a moderate p-value and structureless residuals.

## Even if the best-fit model does pass the ``goodness-of-fit'' test, one still cannot say that the model is the only acceptable model. For example, if the data used in the fit are not particularly good, one may be able to find many different models for which adequate fits can be found. In such a case, the choice of the correct model to fit is a matter of scientific judgment.

- ## Confidence limits of parameters: 
## For a given best-fit parameter ($\theta1$), what is the range of values within which one can consider the true value of the parameter lies? 
That is once we have the best fit model to the data and its best fit parameters, it is now necessary to estimate what are the confidence intervals of the fit parameter values. Lets consider $a$ is our parameter of interest whose confidence interval needs to be determined. The remaining parameters of the model are kept frozen and $a$ is varied between its hard limits (bounds) to obtain different values of the statistic or -log-likelihood.

The confidence interval for a given parameter is computed by varying the parameter value until the statistic increases by a particular amount above the minimum, or ``best-fit'' value. The amount that the statistic is allowed to increase (also referred to as the critical $\Delta$ statistic (Eg: Avni 1976 for $\chi^2$ statistic) depends on the confidence level one requires, and on the number of parameters whose confidence space is being calculated. 

The $\Delta$ is calculated such that probability ($\chi^2$(k degrees of freedom) or -loglikelihood < $\Delta$) = $\alpha$; where $\alpha$ is the confidence level. 
In other words, $\Delta$ is the $\chi^2$ value that one finds from the table of a $\chi^2$ distribution with q degrees of freedom.

# Hypothesis Test and Model Comparison

- ## p-values and Likelihood ratio tests:
When we have two different models that can fit the data well, then we need to choose which model among them that is better. According to Occam's Razor idea of less complex model with a few number of parameters and that reduces the value of the statistic, is generally accepted as the better model. However, this can be more quantified by the methodology of hypothesis testing. Wherein first two exclusive hypothesises are defined such as the null hypothesis ($H_0$) which is the simpler model that is to be rejected and the alternate hypothesis ($H_1$) is another model that is different and more complex than the $H_0$. First, fit the data with both the models, $H_0$ and $H_1$, then obtain the corresponding statistics and note the difference between the statistics, say $\Delta \,\rm Cstat^{*}$. Using the Monte Carlo method, simulate a large number of datasets based on $H_0$, then fit each simulated dataset with both the $H_0$ and $H_1$ models, and calculate the difference in statistics for each case.  
#### Thus, build a cumulative probability distribution plot of the $>\Delta \,\rm Cstat$. Note the value of probability, $p$, of the actual $\Delta \,\rm Cstat^{*}$ on this plot. $(1-p)$ gives the level of confidence of obtaining any value $> \Delta \,\rm Cstat^{*}$. 
If the prescribed significance level of the alternate model, $H_1$ is decided at say $95\%$ (that is, in other words, the chosen significance level is $\alpha=0.05$), then if the value of p is lesser than 0.05, i.e $(1-p) \ge 95\%$, we say **"We reject the null hypothesis $H_0$ at a confidence level of $\ge 95\%$.**  If the p value is greater than 0.05, then we say **"We fail to reject the null hypothesis at the 95% confidence level."**



 

## Note: The term significance level (alpha) is used to refer to a pre-chosen probability (represents the probability of generating false positives; that is, incorrectly rejecting the null hypothesis) and the term "p value" is used to indicate a probability that you calculate after a given study.


## When the p-value falls below the chosen alpha value, then we say the result of the test is statistically significant.


# Threshold based errors - Type I and Type II errors
Statistical testing of a hypothesis leads to two types of error: wrongly rejecting the null hypothesis (Type 1 errors or false positives) and wrongly failing to reject the null hypothesis (Type 2 errors or false negatives). It is impossible to bring these two errors simultaneously to negligible values. 

### Type I error
Here one wrongly rejects the null hypothesis H0 giving a false positive decision. For example, when searching for a faint signal in noise, this occurs when we incorrectly infer that a signal is present when it truly is not.
The probability of committing a type I error (rejecting the null hypothesis when it is actually true) is called $\alpha$ the other name for this is the level of statistical significance.

### Type II error 
Here one wrongly fails to reject the null hypothesis when the alternative is true, giving a false negative decision. In our example, we would incorrectly infer that a signal is absent when it truly is present.
In other words, failing to reject the null hypothesis when it is actually false is referred to as the Type II error. The probability of the occurance of the Type II error is called $\beta$. 




(Bhandari, P. (September 2, 2022). Type I & Type II Errors | Differences, Examples, Visualizations. Scribbr.)


It is important to note that the null hypothesis and the alternative hypothesis are not treated symmetrically. We can only reject the null hypothesis at a given level of significance; we can never accept a null hypothesis. In the case of signal detection, rejecting the null hypothesis leads to detecting a signal. 



If our statistic falls in the critical region of this curve (shaded region), they are considered statistically significant and the null hypothesis is rejected. However, if the null hypothesis is actually true, then this represents the Type I error rate or false positive detection rate or False Discovery Rate.

## There is a trade off between Type I and Type II errors 

The ideal situation would have been if we could minimise both these errors. However, it is not possible. 

Type I and Type II errors occur where these two distributions overlap. The blue shaded area represents alpha, the Type I error rate, and the green shaded area represents beta, the Type II error rate.



Generally, we keep the Type I error to minimum, that is try to reduce the false positive decision. That is we try to reduce the chance of falsely rejecting the null hypothesis to the minimum. This we do by choosing a smaller $\alpha$ value.  

Having said that, it is the scientist who needs to prioritise what are the goals of the test. 

For example, let us say we are very much interested in finding if there is a signal even though the discovery significance level ($\alpha$) in the community is very low (that is larger sigma in the distribution of the null hypothesis). In other words, we don't want to take a chance to miss if the signal is actually there. 
In such a scenario, we will be interested in reducing $\beta$ which in turn will be increasing $\alpha$, thereby the $1-\beta$ that is the probability of the acceptance of the detection increases. 


- ## Upper limits 

Consider a case where NC = 20 photons are counted in a region that is believed to contain a source along with a background. Suppose further that NB = 100 counts are collected in a source-free area $\rho = 10$ times the area of the 
source region. The expected background photons under the source region is $\hat{\lambda_B} = N_B/\rho = 10$, 
and the estimated source strength, $\hat{\lambda_S} = N_C - \hat{\lambda_B} = 10$ and the error on the signal strength which is nearly equivalent to the noise (in an actual scenario the noise would include dark noise, read out noise of the detector etc) is given by $\hat{\sigma_s} = \sqrt{N_C + N_B/\rho^2} \sim 4.6$.  
Now, here the source can be considered undetected by either the signal to noise criterion $\hat{\lambda_S}/\hat{\sigma_s} < 3$ where signal-to-noise threshold is predecided at 3 sigma (predecided alpha).

Or it can be decided on the basis of the value of $\beta$.




In addition to the detection threshold $\alpha$, we require that an existing source should also be detected with probability $1- \beta$, the upper limit to the source brightness is that value which achieves a statistical power of $1-\beta$. A source with higher brightness would be detected more often than $1-\beta$, and vice versa. Thus, specifying an upper limit requires two significance levels, and both the Type I and Type II errors are needed.

# Akaike Information Criterion 
# \begin{equation*}
AIC = 2 k - 2 ln ({\cal{L}}) + \frac{2 k^2 + 2k}{n-k-1}
\end{equation*}

Given a set of candidate models for the data, the preferred model is the one with the minimum AIC value. Thus, AIC rewards goodness of fit (as assessed by the likelihood function), but it also includes a penalty that is an increasing function of the number of estimated parameters. The penalty discourages overfitting, which is desired because increasing the number of parameters in the model almost always improves the goodness of the fit.
